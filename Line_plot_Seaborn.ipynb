```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
fmri=sns.load_dataset('fmri')    # data load from fmri.csv
fmri.head(20)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=fmri)
```

# Add dashes

```python
sns.lineplot(
    data=fmri,
    x="timepoint", y="signal", hue="event", style="event",
    markers=True, dashes=False
)
```

# palette

```python
sns.lineplot(
    data=fmri,
    x="timepoint", y="signal", hue="event", style="event",
    markers=True, dashes=False
,palette='flare')
```
