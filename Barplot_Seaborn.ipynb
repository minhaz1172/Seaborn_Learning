```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=sns.load_dataset('titanic')
data
```

```python
# plot the bar graph
sns.barplot(x='sex',y='survived',data=data)
```

```python
sns.barplot(x='who',y='survived',hue='class' ,data=data)
```
