```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating a 10 bny 10 matrix pof number
data=np.random.randint(1,100,(10,10))
data
```

```python
# plotting heat map
sns.heatmap(data)
plt.show()
```

```python
# select the range of data to show
sns.heatmap(data,cmap='Blues',vmin=30,vmax=70)    # cmap specify the colors
plt.show()
```

```python
sns.heatmap(data,annot=True,vmin=30,vmax=70)    # cmap specify the colors
plt.show()
```
